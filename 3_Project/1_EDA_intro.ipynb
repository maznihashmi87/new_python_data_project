```python
#importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_IND = df[(df['job_country']=='India') & (df['job_title_short']=="Data Analyst")]
```

```python
to_plot_IND = df_DA_IND['job_location'].value_counts().head(10).to_frame()
sns.barplot(data= to_plot_IND, x='count',y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.set_theme(style='ticks')
sns.despine()
plt.title('Number of jobs per job title')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```

```python
df_DA_US = df[(df['job_country']=='United States') & (df['job_title_short']=="Data Analyst")]
```

```python
to_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=to_plot, x = 'count', y ='job_location', hue='count',palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of jobs per job title')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```

```python
dict_column = {
    'job_work_from_home': 'Work from home offered',
    'job_no_degree_mention':'Degree Required',
    'job_health_insurance': 'Health Insurance Offered'

}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches(12, 5)

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(),labels=[False, True], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

```

```python
to_plot = df_DA_US['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=to_plot, x = 'count', y ='company_name', hue='count',palette='dark:b_r', legend=False)
sns.despine()
plt.title('Count of Comapines for Data Analyst in the US')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```
